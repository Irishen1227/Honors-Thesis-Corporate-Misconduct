```python
df['peer'] = 0
from tqdm import tqdm
for i in tqdm(range(len(df_scandal))):
    year = float(df_scandal.iloc[i,7])
    ind_clu = df_scandal.iloc[i,12]
    subset = df.loc[(df['new_fyear'] >= year) & (df['ind_clu'] == ind_clu)]
    subset = subset.reset_index()
    lst = list(subset['index'])
    df.loc[lst,'peer'] = 1
```

```python
df['DTIC'] = ''
for i in tqdm(range(len(df_scandal))):
    ind_clu = df_scandal.iloc[i,12]
    DTIC = df_scandal.iloc[i,11]
    subset = df.loc[(df['ind_clu'] == ind_clu)]
    subset = subset.reset_index()
    lst = list(subset['index'])
    df.loc[lst,'DTIC'] = DTIC
```
